# Статистическое сопоставление цены биткоина с курсами фиатных активов (USD,EUR,RUR), ценами на нефть.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
BTC = pd.read_csv('/content/drive/MyDrive/prob_theory_and_math_stat/BTC-USD.csv')
OIL = pd.read_csv('/content/drive/MyDrive/prob_theory_and_math_stat/brent_oil.csv')
GOLD = pd.read_csv('/content/drive/MyDrive/prob_theory_and_math_stat/gold.csv')
```

```python
BTC
```

```python
OIL
```

```python
BTC.info()
```

```python
OIL.info()
```

```python
BTC.rename(columns = {'Open' : 'BTC open'}, inplace=True)
```

```python
BTC_open = BTC[['Date', 'BTC open']]
BTC_open
```

```python
OIL.rename(columns = {'Open' : 'OIL open'}, inplace=True)
```

```python
OIL_open = OIL[['Date', 'OIL open']]
OIL_open.info()
```

```python
res = BTC_open.merge(OIL_open, on=["Date"])
res.info()
```

```python
res[['BTC open', 'OIL open',]].corr()
```

Вывод: Мы видим, что корреляция биткоина и нефти низкая, это значит направления движения цены вверх или вниз совпадают не часто.

```python
GOLD
```

```python
GOLD.rename(columns = {'Open' : 'GOLD open'}, inplace=True)
GOLD_open = GOLD[['Date', 'GOLD open']]
GOLD_open.head()
```

```python
res1 = BTC_open.merge(GOLD_open, on=["Date"])
res1.head()
```

```python
res1[['BTC open', 'GOLD open',]].corr()
```

Вывод: Мы видим, что корреляция биткоина и золота высокая, это значит направления движения цены вверх или вниз чаще совпадают.
